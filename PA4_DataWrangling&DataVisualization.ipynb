### ECE Board Exam Problem

```python
#first we have to import related library
import pandas as pd
```

```python
#importing the data frame
Vis_data = 'board2.xlsx'
Dframe = pd.read_excel('board2.xlsx')
Dframe
```

```python
#splicing to find the specified requirements
Instru = Dframe.loc[(Dframe['Track']=='Instrumentation') #setting parameters
&(Dframe['Hometown']=='Luzon')
&(Dframe['Electronics']>70), 
['Name','GEAS', 'Electronics']] #setting which column to display
Instru
```

```python
#Before splicing we have to add another column named Average which is the average grade of the student
Dframe['Average'] = (Dframe['Math']+Dframe['Electronics']+Dframe['GEAS']+Dframe['Communication'])/4
```

```python
#splicing to find Female Names, their track and elec score, from Mindanao, with an average higher or equal to 55

#setting parameters
Mindy = Dframe.loc[(Dframe['Average']>=55)
&(Dframe['Hometown']=='Mindanao')
&(Dframe['Gender']=='Female'), 
['Name','Track', 'Electronics', 'Average']]#setting which column to display
Mindy
```

```python
#To be able to create a visualization we have to import matplotlib
import matplotlib.pyplot as plt

#making the range of the y axis 0 to 100
plt.ylim(0,100)

#naming the axes
plt.xlabel('College Track')
plt.ylabel('Average Grade')

#displayung the bar graph in terms of average with respect to Track
plt.bar(Dframe['Track'], Dframe['Average'])
```

```python
#making the range of the y axis 0 to 100
plt.ylim(0,100)

#naming the axes
plt.xlabel('Gender')
plt.ylabel('Average Grade')

#displayung the bar graph regarding average with respect to the Gender
plt.bar(Dframe['Gender'], Dframe['Average'])
```

```python
#making the range of the y axis 0 to 100
plt.ylim(0,100)

#naming the axes
plt.xlabel('Hometown')
plt.ylabel('Average Grade')

#displayung the bar graph regarding average with respect to the Hometown
plt.bar(Dframe['Hometown'], Dframe['Average'])
```
